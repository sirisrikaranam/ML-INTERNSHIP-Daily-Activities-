# Standardization

```python
s=pd.Series([1000,2000,3000,4000])
s
```

```python
s.mean()
```

```python
s.std()
```

```python
zscore=(s-s.mean())/s.std()
zscore
```

```python
zscore=StandardScaler().fit_transform(s.values.reshape(-1,1))
zscore
```

```python
zscore=(s-s.mean())/s.std(ddof=0)
zscore
```

```python
data={
    "temp":[-20,-30,-40,-60],
    "hum":[100,150,140,130]
}
df=pd.DataFrame(data)
df
```

```python
zscore=StandardScaler().fit_transform(df)
zscoredf=pd.DataFrame(zscore, columns=df.columns)
zscore
```
